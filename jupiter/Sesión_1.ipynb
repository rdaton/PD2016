<b>El ordenador más potente del mundo, el chino Sunway TaihuLight, exhibe una potencia de
cómputo de 93 petaFlops (1 petaFlop = 10 15 Flop). Calcula cuántas operaciones en coma
flotante habrı́a realizado hasta el momento actual dicho ordenador si hubiese estado operando
desde el principio de los tiempos, sabiendo que la edad del universo viene a ser de unos 13700
millones de años. Obtén el resultado como Int, Integer, Double.</b>

```haskell
segEnAnyo x = x * 31536000 -- 365 * 24 * 60 * 60
edadUniverso= 13700 * 10^6 
petaFlop = 10^15
segEnDia x = x * 86400 --(24 * 60 * 60)
prac1a = (petaFlop * 93 * segEnAnyo edadUniverso)

"Int: " ++ show (prac1a::Int)
"Integer: " ++ show (prac1a::Integer)
"Double: " ++ show (prac1a::Double)
```

<b>Calcula cuántos años hay en 10^10 segundos (supón que todos los años tienen 365 dı́as; en otro
momento puedes hacerlo teniendo en cuenta bisiestos).</b>

```haskell
anyosEnSeg x = x `div` segEnAnyo 1
prac1b = anyosEnSeg (10^10)
"Anyos en 10^10 segundos: " ++ show (prac1b :: Integer)
```

<b>Calcula cuántos años enteros, dı́as restantes enteros, horas restantes enteras, minutos restantes
enteros y segundos restantes hay en 10 10 segundos.</b>

```haskell
segRestantesAnyo x = x - segEnAnyo (anyosEnSeg x)
:type segRestantesAnyo
diasEnSeg x = segRestantesAnyo x `div` segEnDia 1
:type diasEnSeg
segRestantesDia x = segRestantesAnyo x - segEnDia (diasEnSeg x) 
:type segRestantesDia
horasEnSeg x = segRestantesDia x `div` 3600
:type horasEnSeg
segRestantesHora x = segRestantesDia x - 3600 * horasEnSeg x
:type segRestantesHora
minutosEnSeg x = segRestantesHora x `div` 60
:type minutosEnSeg
segRestantesMinutos x = segRestantesHora x- 60* minutosEnSeg x
:type segRestantesMinutos

y= 10^10 
yToString="10^10"

unaFunc x = [anyosEnSeg x, diasEnSeg x, horasEnSeg x,minutosEnSeg x,segRestantesMinutos x]
:type unaFunc 

prac1c= unaFunc y

yToString ++ " segundos:"
show (prac1c!!0) ++ " Anyos, " ++ show (prac1c!!1) ++ " Dias, " ++ show (prac1c!!2) ++  " Horas, "
    ++ show (prac1c!!3) ++ " Minutos, "++ show (prac1c!!4) ++ " Segundos. "


```
